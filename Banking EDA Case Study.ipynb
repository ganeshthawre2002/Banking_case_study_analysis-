```python
# Histplot For Value Counts For DIfferent Occupations 
```

```python
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()


import matplotlib.pyplot as plt
import seaborn as sns

# Get categorical columns from the DataFrame
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

if "Occupation" in categorical_cols:
    cols_to_plot = ["BRId", "GenderId", "IAId", "Amount_of_Credit_Cards",
                    "Nationality", "Fee_Structure", "Loyalty_Classification",
                    "Properties_Owned", "Risk_Weighting", "Income Band"]

    for col in cols_to_plot:
        plt.figure(figsize=(10, 6))
        sns.histplot(data=df, x=col)
        plt.title(f'Histogram of {col} Count')
        plt.xlabel(col)
        plt.ylabel("Count")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()



```
